```python
wages2 = wages[0:41] 
# wages2.columns
us_wages = wages2['Annual wages (March CPS)'].values
w2 = list(us_wages)
w2.reverse()
```

```python
w3 = map(lambda x: float(x[1:].replace(',','')), w2)
```

```python
years = [1976 + i for i in range(0,41)]
```

```python
import matplotlib.pyplot as plt
plt.plot(years, w3)
plt.show()

```

```python
us_gdp = gdp[gdp['Country Code'] == 'USA'][16:]['Value'].values
us_gdp
```

```python
plt.plot(years, us_gdp)
plt.show()
```

```python
plt.plot(years, us_gdp)
plt.plot(years, w3)
plt.show()
```

```python
last_gdp = us_gdp[len(us_gdp) - 1]
last_wage = w3[len(w3) - 1]
factor = last_gdp/last_wage
w4 = map(lambda x: factor*x, w3)
```

```python
plt.plot(years, us_gdp)
plt.plot(years, w4)
plt.show()
```
