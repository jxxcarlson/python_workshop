```python
import random

def randomWalk(startingValue, n):
    output = []
    value = startingValue
    while value > 0:
        value = value + 2*random.uniform(-1,1)
        output = output + [value]
    if output[-1] >= 0:
        return output 
    else:
        return output[:-1]
```

```python
rw = randomWalk(4,100); t = [i for i in range(0,len(rw))]
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
abscissa = np.zeros(len(rw)); plt.plot(t, rw); plt.plot(abscissa); plt.show()
```
